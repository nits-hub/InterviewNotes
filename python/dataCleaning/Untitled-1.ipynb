```python
import pandas as pd
df = pd.read_csv('data.csv')
print(df)
```

```python
df.dropna(inplace = True)
print(df.to_string())
```

```python
for x in df.index:
    if df.loc[x, "Maxpulse"] > 150:
        df.loc[x, "Maxpulse"] = 2000

print(df.to_string())
```
